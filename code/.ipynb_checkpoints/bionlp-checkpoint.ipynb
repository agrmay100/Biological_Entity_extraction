```python
import numpy as np
import pandas as pd
import nltk
import glob
import errno
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from sklearn.metrics import f1_score
```

```python
# data extraction from brat data folders and converting it to the conll format into dataframe 
f1 = '/BioNLP-ST_2011_genia_'
f2 = '/BioNLP-ST_2011_Epi_and_PTM_'
f3 = '/BioNLP-ST_2011_bacteria_interactions_'
f4 = '/BioNLP-ST_2011_Bacteria_Biotopes_'
f5 = '/BioNLP-ST_2011_Infectious_Diseases_'

#folder = [f1, f2, f3, f4, f5]
file_ext = ['a1', 'a2', 'txt']
folder = [f1]

def check_len(m,n):      
    if(m < n):
        return 'a2'
    else:
        return 'a1'

def convert_data(data_type_path):
    data_type = {}
    data=[]
    data_type['a1'] = []
    data_type['a2'] = []
    for j in range(len(folder)):
        for i in range(len(file_ext)):
            path = './' + data_type_path + folder[j] + data_type_path + folder[j] + data_type_path + '/*.' + file_ext[i]
            files = glob.glob(path)
            for name in files:
                try:
                    with open(name) as f:
                        data_type.setdefault(file_ext[i],[]).append(f.read())
                except IOError as exc:
                    if exc.errno != errno.EISDIR:
                        raise


            #print(len(data_type[file_ext[i]]))

        if(len(data_type['a1'])==len(data_type['a2'])):
            for i in range(len(data_type['a1'])):
                anno = (data_type['a1'][i]+data_type['a2'][i]).split('\n')
                text = data_type['txt'][i]

                str_num = []
                end_num = []
                tag = []
                for i in range(len(anno)):
                    if(len(anno[i]) < 3):
                        continue
                    else:
                        if(anno[i].split('\t')[0][0]== 'T'):
                            ls = anno[i].split('\t')[1].split()
                            tag.append(ls[0])
                            str_num.append(int(float(ls[1])))
                            end_num.append(int(float(ls[2])))  


                # sorting by starting number str_num
                df = pd.concat([pd.Series(tag),pd.Series(str_num),pd.Series(end_num)],axis=1)
                df.sort_values([1], axis=0, ascending=True, inplace=True)

                # converting again to list
                tag = df.iloc[:,0].tolist()
                str_num = df.iloc[:,1].tolist()
                end_num = df.iloc[:,2].tolist()

                if(len(str_num) > 0):
                    part0 = text[:str_num[0]]
                    data = data + list(zip(nltk.word_tokenize(part0),['o']*len(nltk.word_tokenize(part0))))

                    for i in range(len(str_num)):
                        if(end_num[i]==end_num[-1]):
                            continue

                        part1 = text[str_num[i]:end_num[i]+1]
                        data = data + [(part1,tag[i])]

                        part2 = text[end_num[i]+1:str_num[i+1]]
                        data = data + list(zip(nltk.word_tokenize(part2),['o']*len(part2.split())))

                    part3 = text[end_num[-1]:]
                    data = data + list(zip(nltk.word_tokenize(part3),['o']*len(nltk.word_tokenize(part3))))


#         elif(len(data_type['a1']) < len(data_type['a2'])):
        elif(len(data_type['a1']) != len(data_type['a2'])):

            var = check_len(len(data_type['a1']),len(data_type['a2']))
        
#             for i in range(len(data_type['a2'])):
            for i in range(len(data_type[var])):
                anno = (data_type[var][i]).split('\n')
                text = data_type['txt'][i]

                str_num = []
                end_num = []
                tag = []
                for i in range(len(anno)):
                    if(len(anno[i]) < 3):
                        continue
                    else:
                        if(anno[i].split('\t')[0][0]== 'T'):
                            ls = anno[i].split('\t')[1].split()
                            tag.append(ls[0])
                            str_num.append(int(float(ls[1])))
                            end_num.append(int(float(ls[2])))  


                # sorting by starting number str_num
                df = pd.concat([pd.Series(tag),pd.Series(str_num),pd.Series(end_num)],axis=1)
                df.sort_values([1], axis=0, ascending=True, inplace=True)

                # converting again to list
                tag = df.iloc[:,0].tolist()
                str_num = df.iloc[:,1].tolist()
                end_num = df.iloc[:,2].tolist()

                if(len(str_num) > 0):
                    part0 = text[:str_num[0]]
                    data = data + list(zip(nltk.word_tokenize(part0),['o']*len(nltk.word_tokenize(part0))))

                    for i in range(len(str_num)):
                        if(end_num[i]==end_num[-1]):
                            continue

                        part1 = text[str_num[i]:end_num[i]+1]
                        data = data + [(part1,tag[i])]

                        part2 = text[end_num[i]+1:str_num[i+1]]
                        data = data + list(zip(nltk.word_tokenize(part2),['o']*len(part2.split())))

                    part3 = text[end_num[-1]:]
                    data = data + list(zip(nltk.word_tokenize(part3),['o']*len(nltk.word_tokenize(part3))))

 
    df_mid= pd.DataFrame(data,columns=['Token','Tag'])
    return df_mid
```

```python
train_df = convert_data('train_data')
train_df
```

```python
test_df = convert_data('test_data')
```

```python
# Collecting all index of full-stop and adding a blank row after each of them so that it matches CoNll data format
def add_blank_row(df):
    idx = [0]
    idx_temp = list(df[df['Token']=='.'].iloc[:,1].index)
    idx = idx + idx_temp + [-2]

    br = pd.DataFrame({'Token':[" "],'Tag':[" "]})
    data_df = pd.DataFrame(columns=['Token','Tag'])

    for ix,data in enumerate(idx):
        if ix+1 < len(idx):
            if ix == 0:
                data_df = pd.concat([data_df,df.loc[idx[ix]:idx[ix+1]].append(br,ignore_index=True)],ignore_index=True)
            else:
                data_df = pd.concat([data_df,df.loc[idx[ix]+1:idx[ix+1]].append(br,ignore_index=True)],ignore_index=True)        
        else:
            continue

    cols = data_df.columns.tolist()
    cols = ['Token', 'Tag']
    data_df = data_df[cols]
    return data_df
```

```python
train_df = add_blank_row(train_df)
# dev_df = add_blank_row(dev_df)
```

```python
test_df = add_blank_row(test_df)
```

```python
tags = train_df.Tag.value_counts()
print('Value counts of Tags in the training data {}' .format(tags))
```

### Data Cleaning

```python
#removing stopwords like unimportant tags and some useless syntax 
stop_ = ['(', ')', '.', ':', ';', ',', '']
def data_clean(df):
#     for i in range(len(stop_)):
#         df = df[df['Token'] != stop_[i]]
    
    df = df[~df['Token'].str.contains('\n')]
    
    #getting entities for top five tags
    tag = (df.Tag.value_counts())[5:]
    for i in range(len(tag.keys())):
        df = df[~df['Tag'].str.contains(tag.keys()[i])]
    
    return df
```

```python
train_df1 = data_clean(train_df)
test_df1 = data_clean(test_df)
# dev_df1 = data_clean(dev_df)
```

```python
df = pd.read_csv('./data/test_data.tsv', sep = '\t')
```

```python
test_actual = df.iloc[:, 1].ravel()
test_predict = df.iloc[:, 2].ravel()
```

```python
labels = pd.Series(test_predict).value_counts().keys()
```

```python
cm = confusion_matrix(test_actual, test_predict, labels=labels)
```

```python
#plot confusion matrix with seaborn heatmap
import seaborn as sns
corrmat = cm
f, ax = plt.subplots(figsize=(5, 5))
sns.heatmap(corrmat, square=True, annot=True, fmt='.2f', cmap = "summer", xticklabels=labels, yticklabels=labels);
plt.title('Confusion matrix of the classifier')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
#print f1 score with average
print(f1_score(test_actual, test_predict, labels = labels, average  = 'micro'))
```

```python
#print f1 score for individual labels
x = labels
y = f1_score(test_actual, test_predict, labels = labels, average  = None)
pd.DataFrame(list(zip(x, y)), columns = ['Entity', 'f1 score'])
```
